#Linear Regression with Multiple Variables = **Multivariate Regression**

```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

```python
bedroom_median = df.bedrooms.median()
bedroom_median
```

```python
df.bedrooms.fillna(bedroom_median, inplace = True)
df
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df.drop(['price'], axis = "columns"), df.price)

#from sklearn import linear_model
#model = linear_model.LinearRegression()
```

```python
model.coef_
```

```python
model.intercept_
```

```python
mod = LinearRegression()
mod.fit(df[['area','bedrooms','age']], df.price)
```

```python
mod.coef_
```

```python
mod.intercept_
```

```python
model.predict([[3000, 3, 60]])
```

#Hiring.csv - Practice

```python
d = pd.read_csv("hiring.csv")
d
```

```python
pip install word2number
```

```python
from word2number import w2n
```

```python
d.experience.fillna("0", inplace = True)
d
```

```python
d.experience = d.experience.apply(w2n.word_to_num)
d
```

```python
import math
d['test_score(out of 10)'].fillna(math.floor(d['test_score(out of 10)'].mean()), inplace = True)
d
```

```python
d['test_score(out of 10)'].fillna(d['test_score(out of 10)'].median(), inplace = True)
d
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(d.drop(['salary($)'], axis = "columns"), d['salary($)'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print(model.predict([[2, 9, 6]]))
print(model.predict([[12, 10, 10]]))
```
