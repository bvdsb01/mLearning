#Practice - Linear Regression - Canada_Per_Capita_Income.csv and Homeprices.csv

```python
import pandas as pd
df = pd.read_csv("homeprices (2).csv")
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.scatter(df.area, df.price, marker = "+", color = 'red')
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df[['area']], df.price) #only first argument has to be 2D array
```

```python
model.predict([[3300]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print(model.coef_ * 3300 + model.intercept_) # y = mx + b (mathematical equation)
```

```python
df_1 = pd.read_csv("areas.csv")
df_1.head()
```

```python
p = model.predict(df_1[['area']])
p
```

```python
df_1['prices'] = p
df_1.head()
```

```python
df_1.to_csv("prediction.csv", index = False)
```

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.area, df.price, marker = "+", color = 'red')
plt.plot(df.area, model.predict(df[['area']]), color = 'blue')
```

```python
plt.plot(df.area, model.predict(df[['area']]), color = 'blue')
```

#Exercise - Predicted Income in 2020

```python
d = pd.read_csv("canada_per_capita_income.csv")
print(d.shape)
d.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.scatter(d.year, d['per capita income (US$)'], marker = "+", color = "red")
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
```

```python
plt.plot(d.year, d['per capita income (US$)'], marker = "+", color = "blue")
```

```python
#from sklearn.linear_model import LinearRegression
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(d[['year']], d['per capita income (US$)'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[2020]])
```

```python
plt.scatter(d.year, d['per capita income (US$)'], color = 'red', marker = "+")
plt.plot(d.year, model.predict(d[['year']]), color = "purple")
```

```python
model.predict(d[['year']])
```

```python
model.score(d[['year']], d['per capita income (US$)'])
```

```python
model.coef_*1971 + model.intercept_
```
