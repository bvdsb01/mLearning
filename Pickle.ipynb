#Save and load trained model using

- pickle

- sklearn joblib

image.png

Pickle module allows to serialize python object into a file.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("homeprices.csv")
df
```

```python
model = LinearRegression()
model.fit(df[['area']], df.price)
```

```python
model.predict([[5000]])
```

```python
import pickle
```

```python
with open("model_pickle","wb") as f:
  pickle.dump(model,f)
```

```python
with open("model_pickle", "rb") as f:
  mp = pickle.load(f)
```

```python
mp.predict([[5000]])
```

If model consists of large numpy array then using joblib might be more efficient.
